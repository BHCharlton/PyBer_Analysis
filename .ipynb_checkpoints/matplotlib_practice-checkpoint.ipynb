```python
%matplotlib inline
```

```python
#import dependencies
import matplotlib.pyplot as plt
```

```python
#Set the x-axis to represent each month of the year
x_axis = ['Jan', 'Feb', 'Mar', 'April', 'May', 'June', 'July', 'Aug', 'Sept', 'Oct', 'Nov', 'Dec']
#Set the y-axis to represent the total fare in usd for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#Create the plot; USE THIS METHOD
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
#Create the plot and add a legend (also create a label and format)
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label = 'Boston')
plt.legend()
#Create the x and y axes labels
plt.xlabel('Date')
plt.ylabel('Fare (US$)')
#Set the y limit to 45
plt.ylim(0, 45)
#Create a title
plt.title('PyBer Fare by Month')
```

```python
#Format the plot using object-oriented interface
#Create a single x, y axes
fig, ax = plt.subplots()
#Create and format the plot
ax.plot(x_axis, y_axis, marker="d", color="green", linewidth=2, label = 'Boston')
#Add title and axes labels
ax.set_title('PyBer Fare by Month')
ax.set_xlabel('Date')
ax.set_ylabel('Fare (US$)')
#Add legend
ax.legend()
#Set y limit
ax.set_ylim(0, 45)
#Add grid lines
ax.grid()
```

```python
#BAR TIME:Create the plot using same data
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare (US$)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
#Now create a horizontal bar graph
plt.barh(x_axis, y_axis)
#Chain the gca(get current axes) method to the invert_yaxis() to list months in ascending order
plt.gca().invert_yaxis()
```

```python
#Skill Drill
#Create the plot and format the data
plt.barh(x_axis, y_axis, color = "magenta", label = 'Boston')
#Add title and axes labels
plt.title('PyBer Fare by Month')
plt.xlabel('Fare (US$)')
plt.ylabel('Date')
#invert the y-axis to start with Jan on top
plt.gca().invert_yaxis()
#Add legend
plt.legend()
```

```python
#Create a plot using object-oriented method (ax.barh)
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
```

```python
fig, ax = plt.subplots()
ax.barh(y_axis, x_axis)
```

```python
#Skill Drill
#Create the plot and format the data
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color = 'cyan', label = 'Chicago')
#Insert title and axes labels
ax.set_title('PyBer Fare by Month')
ax.set_xlabel('Fare (US$)')
ax.set_ylabel('Date')
#Invert y-axis so Jan is on top
ax.invert_yaxis()
#Add legend
ax.legend()
```

```python
#Scatter Plots using MATLAB method
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(x_axis, y_axis)
```

```python
#Skill Drill
plt.scatter(y_axis, x_axis, color='red', label='Chicago')
plt.title('PyBer Fare by Month')
plt.xlabel('Fare (US$)')
plt.ylabel('Date')
#Set a limit for the x-axis
plt.xlim(0, 50)

plt.gca().invert_yaxis()
```

```python
#Bubble plots (use to add more visual factors)
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
#Change the size of the markers
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
#Create scatter plot using the Object-oriented interface method
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
#Skill Drill
fig, ax = plt.subplots()
#Change size of markers
y_axislarger = []
for data in y_axis:
    y_axislarger.append(data*5)
#Create the plot and change then color of the markers
ax.scatter(y_axis, x_axis, color='skyblue', edgecolor='black', label='Boston', s=y_axislarger)
ax.legend()
#Add titles
ax.set_xlabel('Fare (US$)')
ax.set_ylabel('Date')
ax.set_title('PyBer Fare by Month')
#Set a limit for the x-axis
ax.set_xlim(0, 50)
#Invert the y-axis
ax.invert_yaxis()
```

```python
#Pie Charts using matlab method
#Change size of markers
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
plt.subplots(figsize = (8,8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show()
```
